```python
import pandas as pd

bank = pd.read_csv('word_bank.csv')

bank
```

```python
bank = bank[bank['word'].str.len() >= 3]

bank
```

```python
def isInWordBank(word):
    words = bank[bank['word'].str.upper() == word.upper()]
    return len(words) > 0

def test(word):
    print(f'{word} is {'NOT ' if not isInWordBank(word) else ''}in the word bank')
```

```python
for word in ['fly', 'spoon', 'burger', 'a', 'i', 'as', 'few', 'american']:
    test(word)
```
